### Create a bar chart or histogram to visualize the distribution of categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the csv file
data=pd.read_csv('heart.csv')
```

```python
# The histagram of the Age variable was reached.
data["age"].hist(edgecolor = "black");
```

```python
# Here, 0 - Healthy Individual 1 - Heart-Disease Patient
sns.histplot(data=data, x='age', hue='target', bins=30, kde=True, palette='Set1')
plt.legend([0,1]) 
plt.title('Age Distribution')
plt.show()
```

```python
sns.barplot(x='sex',y='age',data=data,palette="GnBu")
```

```python
data.corr()
```

```python
#get correlations of each features in dataset
corrmat = data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corrmat, cbar=True, square=True, fmt='.1f',annot=True, annot_kws={'size':8}, cmap="Blues")
```
